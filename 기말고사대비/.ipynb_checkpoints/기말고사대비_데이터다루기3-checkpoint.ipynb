```python
import csv
f = open('seoul(19040101_20190117).csv')
data = csv.reader(f)
next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month =='08':
            aug.append(float(row[-1]))

import matplotlib.pyplot as plt
#plt.figure(dpi = 300)
plt.hist(aug, bins = 100, color = 'r')
plt.show()
```

```python
import csv
f = open('seoul(19040101_20190117).csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month =='08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

import matplotlib.pyplot as plt
#plt.figure(dpi = 300)
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```

```python
import csv
f = open('seoul(19040101_20190117).csv')
data = csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
#plt.figure(figsize(10, 5), dpi = 300)
plt.boxplot(month)
plt.show()
```

```python
import csv
f = open('seoul(19040101_20190117).csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.boxplot(day, showfliers = False)
plt.show()
```

```python
import matplotlib.pyplot as plt
f = open('seoul(19040101_20190117).csv')
data = csv.reader(f)
next(data)

day = [[] for i in range(31)]

for row in data:
    if row[-1] != '' :
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.style.use('ggplot')
plt.figure(figsize = (10, 5), dpi = 300)
plt.boxplot(day, showfliers = False)
plt.show()
```
