```python
df2 = df[df['소재지(지번)'].str.contains('성거읍')]
df2
```

```python
df3 = df[df['소재지(지번)'].str.contains('입장면')]
df3
```

```python
df = pd.DataFrame(
    {'지역별':['성환읍', '성거읍', '입장면'],
    '음식점 수': [ 371, 206, 157 ]})
df['순위']=df['음식점 수'].rank(method='dense',ascending=False)
df
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']='NanumGothic'

ratio = [371, 206, 157]
labels = ['성환읍', '성거읍', '입장면']

plt.pie(
    ratio,
    labels = labels,
    autopct = '%.1f%%',
    startangle = 250,
    counterclock = False
)
plt.show()
```
