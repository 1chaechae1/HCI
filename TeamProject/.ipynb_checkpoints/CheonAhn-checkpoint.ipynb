```python
df=pd.DataFrame({'연령구간':['10~19세','20~29세','30~39세','40~49세','50~59세','60~69세','70~79세','80~89세','90~99세',
                            '100세 이상'],'인구수':[67666.0,95503.0,102519.0,111471.0,103669.0,69259.0,30619.0,15764.0,2324.0,84.0]})
df['순위']=df['인구수'].rank(ascending = False)
df
```

```python
Ten=df[df['연령구간']=='10~19세']['인구수']
Twenty=df[df['연령구간']=='20~29세']['인구수']
Thirty=df[df['연령구간']=='30~39세']['인구수']
Forty=df[df['연령구간']=='40~49세']['인구수']
Fifty=df[df['연령구간']=='50~59세']['인구수']
Sixty=df[df['연령구간']=='60~69세']['인구수']
Seventy=df[df['연령구간']=='70~79세']['인구수']
Eighty=df[df['연령구간']=='80~89세']['인구수']
Ninety=df[df['연령구간']=='90~99세']['인구수']
Hundred=df[df['연령구간']=='100세 이상']['인구수']
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family']='NanumGothic'
ratio=[Ten.mean(),Twenty.mean(),Thirty.mean(),Forty.mean(),Fifty.mean(),Sixty.mean(),Seventy.mean(),Eighty.mean(),Ninety.mean(),
       Hundred.mean()]

labels=['10~19세','20~29세','30~39세','40~49세','50~59세','60~69세','70~79세','80~89세','90~99세','100세 이상']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
