```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'

categories = ['/', '', '/', '', '', '/']
categories = [*categories, categories[0]]

dongnamgu = [11, 5, 13, 10, 4, 13]
seobukgu = [0, 2, 0, 1, 2, 5]
dongnamgu = [*dongnamgu, dongnamgu[0]]
seobukgu = [*seobukgu, seobukgu[0]]

label_loc = np.linspace(start = 0, stop = 2*np.pi, num = len(dongnamgu))

plt.figure(figsize = (12, 12))
ax = plt.subplot(polar = True)
plt.xticks(label_loc, labels = categories, fontsize = 13)
ax.plot(label_loc, dongnamgu, label = '', linestyle = 'dashed', color = 'lightcoral')
ax.fill(label_loc, dongnamgu, color = 'lightcoral', alpha = 0.3)
ax.plot(label_loc, seobukgu, label = '', linestyle = 'dotted', color = 'skyblue')
ax.fill(label_loc, seobukgu, color = 'skyblue', alpha = 0.3)
ax.legend()
plt.show()
```
